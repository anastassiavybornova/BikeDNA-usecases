### Preparing input data for BikeDNA analysis of Belgrade, Serbia.

So far we have not found reference data in digitized format, so analysis is ran only for OpenStreetMap data (only intrinsic part of analysis). The only input file we need to prepare here is a gpkg file containing the polygon of the city area.

```python
# load libraries
import osmnx as ox
import geopandas as gpd
```

```python
# get city boundaries
gdf = ox.geocode_to_gdf("Belgrade, Serbia")
```

```python
# convert to projected CRS
gdf = gdf.to_crs("EPSG:8682")
```

```python
gdf.plot()
```

```python
gdf.to_file("../data/study_area_polygon/Belgrade/study_area_polygon.gpkg", index = False)
```
